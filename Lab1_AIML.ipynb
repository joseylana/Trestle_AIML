```python
import pandas as pd
import numpy as np

import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load titanic dataset
titanic_dataset = sb.load_dataset('titanic')
```

```python
titanic_dataset
```

```python
# identifying missing data
pd.isnull(titanic_dataset)
```

```python
# dropping all dataset with NaN as input
new = titanic_dataset.dropna()
```

```python
# plot in matplotlib
plt.hist(new['age'], color='orange')
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
# plot in seaborn
sb.histplot(data=new, x="age")
plt.title('Age Distribution of Passengers')
```

```python
# boxplot in seaborn
sb.boxplot(data=new, x="survived", y="fare")
```

```python
# boxplot in matplotlib
new.boxplot(by ='survived', column =['fare'], grid = False)
plt.show()
```

```python
# heatmap
sb.heatmap(new.corr())
```

```python
sb.pairplot(new)
plt.show()
```
